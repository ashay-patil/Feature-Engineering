## 5 Number Summary

```python
import numpy as np
```

```python
lst = [-150, 45, 32, 56, 75, 89, 90, 87, 67, 54, 45, 98, 99, 67, 74, 200]
```

```python
min, Q1, median, Q3, max = np.quantile(lst, [0,0.25, 0.5, 0.75, 1])

min, Q1, median, Q3, max 
```

```python
IQR = Q3-Q1
IQR
```

```python
lower_fence = Q1 - 1.5 * IQR
higher_fence = Q3 + 1.5 * IQR

lower_fence, higher_fence
```
