## Target Guided Oridinal Encoding

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'city' : ['New York', 'London', 'Paris' , 'Tokyo', 'New York', 'Paris'],
    'price' : [200, 150, 300, 250, 180, 320]
})

df
```

```python
grouped_df = df.groupby('city')['price'].mean()

grouped_df
```

```python
mean_price_dict = grouped_df.to_dict()

mean_price_dict
```

```python
df['encoded_city'] = df['city'].map(mean_price_dict)

df
```

## Practice Exercise 

```python
import seaborn as sns

df = sns.load_dataset('tips')

df.head()
```

```python
grouped_time = df.groupby('time')['total_bill'].mean()

mean_time_dict = grouped_time.to_dict()

df['encoded_time'] = df['time'].map(mean_time_dict)

df.head()
```
