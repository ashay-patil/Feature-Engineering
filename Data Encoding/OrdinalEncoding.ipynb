## Ordinal Encoding

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'size' : ['small', 'medium', 'large', 'medium', 'small', 'large']
})

df
```

```python
from sklearn.preprocessing import OrdinalEncoder

ordinal_encoder = OrdinalEncoder(categories=[['small', 'medium', 'large']])
```

```python
ordinal_encoded = ordinal_encoder.fit_transform(df[['size']])

ordinal_encoded
```

```python
ordinal_encoded_df = pd.DataFrame(ordinal_encoded, columns=['encoded_size'])

ordinal_encoded_df
```

```python
final_df = pd.concat([df, ordinal_encoded_df], axis=1)

final_df
```

```python
new_val = [['small']]
encoded_val = ordinal_encoder.transform(new_val)
encoded_val
```
