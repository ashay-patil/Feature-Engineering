## Label Encoding

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'colors' : ['red', 'blue', 'green', 'green', 'red', 'blue']
})

df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

lbl_encoder = LabelEncoder()
```

```python
lbl_encoded = lbl_encoder.fit_transform(df[['colors']])

lbl_encoded
```

```python
lbl_encoded_df = pd.DataFrame(lbl_encoded, columns=['encoded_colors'])

lbl_encoded_df
```

colors will be assigned the numbers as per their alphabetical order 

red -> 2 <br>
blue -> 0 <br>
green -> 1


The numbers will range from 0 to total_categories - 1

```python
final_df = pd.concat([df, lbl_encoded_df], axis=1)

final_df
```

```python
new_val = [['blue']]

encoded_val = lbl_encoder.transform(new_val)

encoded_val
```
