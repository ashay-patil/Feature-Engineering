### Deleting Rows or Columns with Missing Values

```python
df.shape 
```

There are 891 rows and 15 columns in the dataframs

One way to handle missing values is to delete the rows or columns containing them. However, this can result in significant data loss.

```python
df.dropna().shape
```

df.drop() returns a new dataframe in which all the rows having any missing value are deleted. We can observe that initially there were 891 rows and after deleting rows with missing value we are left with 182 rows.... this is a significant loss of data. If a column has a large number of missing values, it may be suitable to delete that column.  From df.isnull().sum() we gotta know that deck is having 688 missing values and total rows are 891... thus it is better to entirely delete that column

```python
newDf = df.dropna(axis=1)
```

df.dropna(axis = 1) drops all the columns with any missing value and returns a new dataframe.... In df we have 4 Columns having empty values (age, embarked, deck and embark_town). After doing df.dropna(axis=1) all these columns got deleted
