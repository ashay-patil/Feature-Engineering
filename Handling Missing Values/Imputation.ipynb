## Imputation Techniques for Missing Values

Instead of deleting data, imputation techniques can be used to fill in missing values. The main imputation techniques are mean, median, mode, and random sample imputation.

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('titanic')
```

### Mean Value Imputation 

Mean value imputation replaces missing values with the mean of the column. This technique works well when the data is normally distributed.
